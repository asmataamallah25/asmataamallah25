```python
import pandas as pd
import numpy as np

#sklearn is a machine learning library
#train_test_split is a function used to split the dataset into training and test subsets
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('/kaggle/input/bank-note-authentication-uci-data/BankNote_Authentication.csv')
```

```python
dataset
```

```python
#dependant and independant features
#[:,:-1] is a sclicing notation that means selecting all raws and columns except for the last column
x = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
x.head(5)
```

```python
y.head(3)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
#implement random forest classifier
from sklearn.ensemble import RandomForestClassifier
Classifier = RandomForestClassifier()
Classifier.fit(x_train, y_train)
```

```python
#prediction
y_predict = Classifier.predict(x_test)
```

```python
#check accuracy
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test, y_predict)
score
```

```python
import pickle
pickle_out = open ('Classiifer.pkl', 'wb')
pickle.dump(Classifier, pickle_out)
pickle_out.close()
```
